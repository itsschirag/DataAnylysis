```python
y= data.Price
```

```python
X= data.drop(['Price'], axis=1)
```

```python
X
```

```python
X.tail()
```

```python
X=X.drop(['Date'],axis=1)
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OrdinalEncoder,OneHotEncoder,StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer 
```

```python
X.columns
```

```python
cat_var=['Suburb','Address','Type','Method','SellerG', 'CouncilArea', 'Regionname']
```

```python
get_num_var=X.drop(cat_var,axis=1)
num_var=list(get_num_var.columns)
num_var
```

```python
num_pipeline=Pipeline([('sim',SimpleImputer(strategy='median')),
                  ('std_scaler', StandardScaler())])
transformer=ColumnTransformer([
    ('num',num_pipeline,num_var),
    ('cat',OrdinalEncoder(),cat_var)
])

prepared=transformer.fit_transform(X)
```

```python
prepared=pd.DataFrame(prepared)
```

```python
prepared
```

```python
prepared.tail()
```

```python
prepared.columns=X.columns
```

```python
prepared
```

```python
prepared=prepared.drop(['Regionname'],axis=1)
```

```python
prepared
```

```python
from sklearn.model_selection import train_test_split

xtrain, xvalid, ytrain, yvalid=train_test_split(prepared,y,test_size=0.2)
```

```python
xtrain
```

```python
ytrain
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error

rand=RandomForestRegressor()
rand.fit(xtrain,ytrain)

ypredict=rand.predict(xvalid)
for_mse= mean_squared_error(yvalid,ypredict)
rmse=np.sqrt(for_mse)
rmse
```

```python
ytrainpred=rand.predict(xtrain)
msefor_overfitting=mean_squared_error(ytrainpred,ytrain)
rsme_over=np.sqrt(msefor_overfitting)
rsme_over
```

```python
from sklearn.tree import DecisionTreeRegressor

tree=DecisionTreeRegressor()
tree.fit(xtrain,ytrain)
ytpredict=tree.predict(xvalid)
mse_tree=mean_squared_error(ytpredict,yvalid)
rsme_tree=np.sqrt(mse_tree)
mse_tree
```

```python
scores=cross_val_score(tree,xtrain,ytrain,cv=10,scoring='neg_mean_squared_error')

def scoring(scores):
    print("scores:",scores)
    print("score_meaan:",scores.mean())
    print("scores_std:", scores.std())

scoring(-scores)
```
